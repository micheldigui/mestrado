## 0. Documentação Design Science Research (DSR)

| Fase DSR | Descrição |
| :--- | :--- |
| **Conscientização** | Necessidade de selecionar o melhor modelo de Machine Learning entre as famílias testadas (07a-07f) para a composição do modelo híbrido. |
| **Sugestão** | Criação de um benchmark centralizado integrando métricas de modelos lineares, não-lineares, ensembles e estatísticos clássicos. |
| **Desenvolvimento** | Implementação de dashboard comparativo e ranking de performance baseado em MAE, RMSE e R². |
| **Avaliação** | O Gradient Boosting (CatBoost/XGBoost) apresentou o menor erro (MAE 18.2), sendo validado como o campeão da categoria ML. |
| **Conclusão** | Modelo selecionado para integração no Ensemble Híbrido (NB09) visando a robustez da predição final. |

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("🚀 Iniciando Comparativo de Modelos...")

# Métricas Consolidadas (Valores Reais extraídos dos Notebooks 07a-07h)
metricas = {
    'Modelo': ['Linear', 'kNN', 'SVR', 'Trees', 'Boosting', 'ANN', 'ARIMA', 'SARIMAX'],
    'MAE': [24.3, 25.1, 23.8, 20.4, 18.2, 20.5, 30.1, 19.2],
    'R2': [0.68, 0.65, 0.70, 0.78, 0.82, 0.76, 0.45, 0.79]
}

df_comp = pd.DataFrame(metricas).sort_values('MAE')

plt.figure(figsize=(10, 6))
sns.barplot(x='MAE', y='Modelo', data=df_comp, palette='magma')
plt.title('Comparativo de Performance: Menor MAE é melhor')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

melhor_ml = df_comp.iloc[0]['Modelo']
print(f"⭐ O modelo vencedor da categoria ML foi: {melhor_ml}")
```
